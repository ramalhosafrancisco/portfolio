#### **Q7:**
Das variáveis numéricas, qual(is) parece(m) ajudar melhor a identificar os clientes que abandonam? (palavra-chave: *pairplot*)

#### **Código:**

```python
variaveis_numericas = ['tenure', 'MonthlyCharges', 'TotalCharges']

ax=sns.pairplot(cust_data[[*variaveis_numericas, 'Churn']], hue='Churn')
ax.fig.suptitle(f"Relação entre Variáveis Numéricas e 'Churn'", y=1.03)
plt.show()
```

#### **Resposta:**

As variáveis 'TotalCharges' e 'ternure', apresentam uma correlação positiva forte (0,83). Pela visualização do gráfico 'pairplot' é possível verificar a existência de uma linha reta inclinada para cima da esquerda para a direita, indicando que à medida que o tempo de permanência aumenta ('ternure'), o 'TotalCharges' também aumenta de forma linear.  
A coloração laranja (cor que corresponde ao cancelamento do serviço por parte do cliente) encontra-se maioritariamente agrupada ao longo da diagonal do gráfico, sugerindo que a relação entre estas variáveis numéricas pode estar associada à variável 'Churn'.  
Clientes com elevados valores de 'TotalCharges' tem maior tendência em abandonar a empresa no decorrer dos 60 primeiros meses comparativamente aqueles que tem valores de 'TotalCharges' mais baixos para o mesmo intervalo de tempo. Pela observação do 'pairplot' é ainda possível constatar que a partir do 60º mês de adesão, a taxa 'Churn' já não é tão pronunciada como em relação aos meses anteriores (inclusivamente para valores elevados de 'TotalCharges'). 


As variáveis 'TotalCharges' e 'MonthlyCharges', apresentam uma correlação positiva moderada (0,65), e a coloração laranja apresenta-se predominantemente agrupada nos valores mais baixos de 'TotalCharges' e também nos valores mais altos de 'MonthlyCharges'. Isto pode sugerir, que nos primeiros meses de adesão existe elevada incidência de cancelamentos por parte dos clientes, por outro lado, à medida que o tempo de permanecia aumenta, as rescisões começam a ser mais frequentes nos clientes que pagam mensalidades elevadas (mais do que 75($) por mês).  
Nos clientes que tem 'TotalCharges' muito elevados (nos diversos tipos de mensalidade), a taxa Churn é menos pronunciada, o que pode estar relacionado com o tempo que estão na empresa, clientes que permanecem um determinado tempo na empresa (traduzindo-se em elevados valores acumulados de 'TotalCharges') tem menor probabilidade de cancelar o serviço (geralmente a partir do 60º mês de adesão). O 'Churn' é praticamente inexistente nos clientes que pagam os valores de mensalidade mínimos. 

As variáveis 'MonthlyCharges' e 'ternure', apresentam uma correlação positiva fraca (0,25), e coloração laranja apresenta-se mais dispersa ao longo dos gráficos. Pela leitura do 'paiplot' verifica-se uma elevada concentração de desistencias do serviço nos 20 primeiros meses de adesão (em quase todos os valores de cobrança mensal, mas com relativa maior incidência nas cobraças mensais mais elevadas). A partir do 20º mês o número de cancelamentos em termos gerais é mais frequente em valores de mensalidade elevados. Tal como em 'TotalCharges', a partir do 60º mês a taxa 'Churn' já é bastante menos evidenciada do que em relação aos meses anteriores (para todos os valores de 'MonthlyCharges'). Aqui também é possível verificar que o 'Churn' é praticamente inexistente nos clientes que pagam os valores de mensalidade minimos.

Pela análise dos padrões identificados nos gráficos e correlações, considero que as variáveis numéricas que melhor parecem ajudar a identificar os clientes que abandonam a empresa são: 'TotalCharges' e 'ternure'.

#### **Q8:**

Em relação às variáveis categóricas, quais terão maior potencial preditivo? Analise cuidadosamente os resultados.<br/>
(**Sugestão:** Para cada variável categórica, calcule a % de abandonos em cada categoria - preferencialmente com representação gráfica)

#### **Código:**

##### % total de clientes que abandoram a empresa:

```python
total_clientes=cust_data['customerID'].count()
print('Total de clientes: ', total_clientes)
total_clientes_churn_yes=(cust_data['Churn']=='Yes').sum()
print('Total de clientes que abandonaram a empresa: ', total_clientes_churn_yes)
perc_abandonos_total=(total_clientes_churn_yes/total_clientes)*100
print('% total de clientes que abandoram a empresa. ', perc_abandonos_total.round(1))
```

### % de abandonos em cada categoria para cada uma das variáveis categóricas (com representação gráfica):

```python
variaveis_categoricas = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
       'PaymentMethod', 'Churn']

variaveis_categoricas_churn_yes=[]


for variavel in variaveis_categoricas:
    if variavel != 'Churn':
        churn_perc = (cust_data.groupby(variavel)['Churn'].value_counts(normalize=True)*100).unstack()
        churn_perc = churn_perc.rename(index={'Bank transfer (automatic)': 'Bank transfer', # aleterei estes nomes de categorias para tornar a visualização dos gráficos mais fácil.
                                                        'Credit card (automatic)': 'Credit card',
                                                        '0': 'No',
                                                        '1': 'Yes'})
        display(churn_perc.round(1))
        churn_perc.index = [f'{variavel}:\n{cat}' for cat in churn_perc.index] ## associei o nome da variável à categoria tambem para uma melhor interpretação dos graficos.
        variaveis_categoricas_churn_yes.append(churn_perc['Yes'])
        ax=churn_perc.plot(kind='bar', title=f'Taxa Churn - {variavel}')
        ax.set_ylabel('Percentagem')
        ax.set_xticklabels(ax.get_xticklabels(), rotation=0)
        ax.set_ylim(0, 100)
        sns.despine()
        plt.show()

variaveis_categoricas_churn_yes=pd.concat(variaveis_categoricas_churn_yes).sort_values().reset_index()
variaveis_categoricas_churn_yes = variaveis_categoricas_churn_yes.set_index('index')
variaveis_categoricas_churn_yes.index.name = 'Variavel: Categoria'
variaveis_categoricas_churn_yes.index = variaveis_categoricas_churn_yes.index.str.replace('\n', ' ')
variaveis_categoricas_churn_yes = variaveis_categoricas_churn_yes.rename(columns={'Yes': '% Churn'})
display(variaveis_categoricas_churn_yes.round(1).tail(10).sort_values(by='% Churn', ascending=False))

categorias_churn_yes = variaveis_categoricas_churn_yes['% Churn'].astype(float)
variaveis_com_maior_poder_preditivo=categorias_churn_yes[categorias_churn_yes>40].sort_values()
ax=variaveis_com_maior_poder_preditivo.plot(kind='barh', figsize=(8, 6))
for i, v in enumerate(variaveis_com_maior_poder_preditivo):
     ax.text(v, i, f'{v:.1f}%', color='white', ha="right")
ax.set_title(f"Variáveis/Categorias .................................................................................com 'Churn' > 40%", loc='left')
ax.set_ylabel(' ')
ax.set_xlabel('Taxa Churn (%)')
ax.xaxis.set_label_coords(0.7, 0)
ax.tick_params(axis='both', which='both', bottom=False, left=False, labelbottom=False)
ax.set_xlim(0, 70)
sns.despine(left=True, bottom=True)
plt.show()
```

#### **Resposta:**

As variaveis "PaymentMethod", "Contract", "InternetService", "OnlineSecurity", "SeniorCitizen" e "TechSupport" apresentam as categorias com as maiores taxas de abandono de clientes na empresa (>40%), respetivamente, "Eletronic check", "Month-to-month", "Fiber optic", "No Online Security", "Senior" e "No Tech Support".  
Entre 41-46% dos clientes que estavam inseridos em pelo menos uma destas categorias abandonou a empresa. Por esse motivo, as variáveis referidas, podem estar associadas a um maior poder preditivo 'Churn' comparativamente às restantes variáveis.
