### **7.2. Contagem de 'Outliers' para cada Variável (1.5 * IQR)**

```python
# Calculou-se o intervalo interquartil (IQR) para cada característica
Q1 = colunas_numericas_dados.quantile(0.25)
Q3 = colunas_numericas_dados.quantile(0.75)
IQR = Q3 - Q1

# Definiu-se os limites inferior e superior para valores atípicos (outliers) - 3 vezes a distância IQR.
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Contagem de valores 'outliers' para cada característica numérica no DataFrame 'colunas_numericas_dados'.  
# Verificou-se a existencia de valores abaixo do limite inferior ou acima do limite superior, usando condições baseadas no intervalo interquartil (IQR). 
outliers_count = ((colunas_numericas_dados < lower_bound) | (colunas_numericas_dados > upper_bound)).sum()

print("Contagem de outliers para cada variável.")
print(outliers_count)
```

As correlações entre o número de 'caixas tradicionais' e a 'produtividade/hora', e entre o número de 'self checkouts' e a 'produtividade/hora' são:
- Correlação entre o número de 'caixas tradicionais' e produtividade/hora: aproximadamente 0.25;
- Correlação entre o número de 'self checkouts' e produtividade/hora: aproximadamente 0.50.

Esses valores de correlação indicam uma correlação positiva moderada entre o número de 'self checkouts' e a produtividade por hora, e uma correlação positiva fraca entre o número de 'caixas tradicionais' e a produtividade por hora.

Portanto, parece que há uma relação mais forte entre o número de 'self checkouts' e a produtividade por hora do que entre o número de 'caixas tradicionais' e a produtividade por hora.  
No entanto, correlação não implica causalidade, e outros fatores podem estar a influenciar a produtividade por hora das lojas.

#### **9.3.6. Relação entre Número de Colaboradores e a Produtividade/Hora**

```python
correlacao_colaboradores = df['n_colaboradores'].corr(df['produtividade/hora'])

print("Correlação entre o número de colaboradores e produtividade/hora:", correlacao_colaboradores.round(2))
```

A correlação entre o número de colaboradores e a produtividade por hora é de aproximadamente 0.54. Essa correlação positiva moderada sugere que há uma relação significativa entre essas duas variáveis: à medida que o número de colaboradores aumenta, a produtividade por hora tende a aumentar também.

Correlação entre 'itens'/'valor'/'skus' e 'produtividade/hora' também poderiam ser realizadas, no entanto, é razoável pensar que se uma loja vende mais itens, gera mais receita ou tem uma variedade maior de produtos disponíveis, ela possa ser mais produtiva por hora. Como já é esperado que esses fatores contribuam para uma maior produtividade, as correlações entre eles podem ser menos informativas e mais evidentes. Uma alta correlação positiva seria mais uma confirmação do óbvio, e uma correlação negativa seria contraintuitiva. Neste caso optamos por não proceder às correlações mais evidentes referidas.

Essa relação positiva entre a variedade de produtos e as vendas pode ser atribuída à maior oferta disponível para satisfazer as necessidades dos clientes, abrangendo uma variedade mais ampla. Além disso, é importante destacar que a correlação entre a variedade de produtos e a produtividade por hora é também positiva e moderada (0.32), indicando que a variedade de produtos não parece ter um impacto negativo na produtividade ao nível logístico. Estes resultados sugerem que, além de ajudar a subir as vendas, uma maior variedade de produtos pode ser gerenciada de forma relativamente eficiente, sem comprometer as operações em loja.

Isto significa que há uma tendência de aumento na produtividade por hora à medida que a percentagem de caixas self checkouts aumenta, mas a relação não é extremamente forte, ou seja, o aumento na percentagem de self checkouts não causa necessáriamente um aumento direto na produtividade por hora.

Esses resultados indicam que, em média, as lojas com self checkouts têm uma produtividade por hora significativamente maior (261.9) em comparação com as lojas que possuem apenas caixas tradicionais (146.6). Isso sugere que a presença de self checkouts pode estar associada a uma maior eficiência e produtividade no atendimento ao cliente, possivelmente devido à automação e redução do tempo de espera nas filas de pagamento.
