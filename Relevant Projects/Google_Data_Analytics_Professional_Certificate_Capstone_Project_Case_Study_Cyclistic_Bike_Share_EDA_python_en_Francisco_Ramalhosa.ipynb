## Data Cleaning

### Edited headers in 'Trips_2019_Q2' DataFrame

```python
Trips_2019_Q2= Trips_2019_Q2.rename(columns={'01 - Rental Details Rental ID': 'trip_id',
 '01 - Rental Details Local Start Time': 'start_time',
 '01 - Rental Details Local End Time': 'end_time',
 '01 - Rental Details Bike ID': 'bikeid',
 '01 - Rental Details Duration In Seconds Uncapped': 'tripduration',
'03 - Rental Start Station ID': 'from_station_id',
'03 - Rental Start Station Name': 'from_station_name',
'02 - Rental End Station ID': 'to_station_id',
'02 - Rental End Station Name': 'to_station_name',
'User Type': 'usertype',                                             
'Member Gender': 'gender',
 '05 - Member Details Member Birthday Year': 'birthyear'                                            
})
```

### Dataframes Concatenation

To combine the 4 dataframes into a single dataframe I used 'pd.concat()' method.

```python
trips_2019= pd.concat([Trips_2019_Q1, Trips_2019_Q2, Trips_2019_Q3, Trips_2019_Q4], ignore_index=True)
trips_2019.head()
```

```python
trips_2019.info()
```

```python
trips_2019.shape
```

### Column headers formatation and 'user_type' variable category names

Now I have only one dataframe its more simple to do changes in the headers name formating. I will edited headers with a similar formatting (snake_case and lowercase).  
For better understanding and to be allign with the names given to each users category in the 'Scenario' description, I  changed 'user_type' variable category names, from 'Subscriber'and 'Customer' to 'Member' and 'Casual Rider', respectively.

```python
trips_2019= trips_2019.rename(columns={'bikeid': 'bike_id',
 'tripduration': 'trip_duration_seconds',
 'usertype': 'user_type',
 'birthyear': 'birth_year'                                            
})

trips_2019['user_type']=trips_2019['user_type'].replace('Subscriber','Member')
trips_2019['user_type']=trips_2019['user_type'].replace('Customer','Casual Rider')

trips_2019.head()
```

### Missing values

Checked the existence of null values by using the '.isnull()' method.

```python
trips_2019.isnull().sum()
```

Removed the null values by using the '.dropna()' method.

```python
trips_2019 = trips_2019.dropna()
```

```python
trips_2019.info()
```

```python
trips_2019.shape
```

### Duplicates

Checked for duplicates by using '.duplicated()' method.

```python
number_of_duplicates = trips_2019.duplicated().sum()
number_of_duplicates
```

### Extra white spaces in string values

The following 'regular expression' searches for and removes patterns that match consecutive spacings (two or more), spacings at the beginning and end, as well as spacings before or after a hyphen (-). Only columns with string values will be analyzed for each of the DataFrames.

```python
for col in trips_2019.columns:
    if trips_2019[col].dtype == 'object':
        trips_2019[col] = trips_2019[col].str.strip()
        trips_2019[col] = trips_2019[col].str.replace(r'\s{2,}', ' ', regex=True).str.replace(r'\s*-\s*', '-', regex=True)
```

### Add new variables

Added new columns that could be useful for the study.

```python
# Convert 'end_time' and 'start_time' to datetime
trips_2019['end_time'] = pd.to_datetime(trips_2019['end_time'])
trips_2019['start_time'] = pd.to_datetime(trips_2019['start_time'])

# Calculate ride length
trips_2019['ride_length'] = trips_2019['end_time'] - trips_2019['start_time']

# Extract day of the week
trips_2019['day_of_week'] = trips_2019['start_time'].dt.day_name() 

# Calculate user age
trips_2019['age'] = 2019 - trips_2019['birth_year']
```

### Inconsistencies check

I searched for inconsistencies and verified if there are data trip that was registered before 2019 or after 2020 (after 2020, because some users could start the trip in the last days of 2019 and ended during january).

```python
trips_2019['start_year']=trips_2019["start_time"].dt.year
trips_2019['end_year']=trips_2019["end_time"].dt.year
condition_start_year =(trips_2019['start_year']!=2019).sum()
condition_end_year =((trips_2019['end_year'] != 2019) & (trips_2019['end_year'] != 2020)).sum()
print(condition_start_year)
print(condition_end_year)
```

```python
trips_2019[trips_2019['end_year'] == 2020]
```

As can be observed, there are no inconsistencies regarding the 'start_time' and 'end_time' dates. Trips that ended in January 2020 were almost all started during the last days of 2019 (29th and 31st of December) and ended in the first two days of January 2020 (only one user (trip_id: 25811874) ended the trip later, on 10th of January).

Also checked for the existence of trips that have the inconsistency of ending before they started (possibly some system error during 2019-11-03).  
I opted to remove these observations.

```python
trips_2019[trips_2019['end_time'] < trips_2019['start_time']]
```

```python
trips_2019 = trips_2019[trips_2019['end_time'] >= trips_2019['start_time']]
```

Removed variables that would not be relevant to the study or whose information is already included in another variable.

### Start and End Trip Time (Hour of the Day)

```python
hour_counts_start = trips_2019['start_time'].dt.floor('H').dt.hour.value_counts()

print("Count of Started Trips for each hour:")
print(hour_counts_start)
```

```python
hour_counts_end = trips_2019['end_time'].dt.floor('H').dt.hour.value_counts()

print("Count of Ended Trips for each hour:")
print(hour_counts_end)
```

```python
# Create subplots with one row and two columns
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

# Sort the hour counts in descending order for started trips
hour_counts_sorted = hour_counts_start.sort_values(ascending=False)

# Plot for started trips
axes[0].bar(hour_counts_sorted.index, hour_counts_sorted.values, color='green')
axes[0].set_xlabel('Hour')
axes[0].set_ylabel('Count')
axes[0].set_title('Count of Started Trips for Each Hour')
axes[0].tick_params(axis='x') 
plt.sca(axes[0])
plt.xticks(hour_counts_sorted.index)

# Sort the hour counts in descending order for ended trips
hour_counts_sorted_end = hour_counts_end.sort_values(ascending=False)

# Plot for ended trips
axes[1].bar(hour_counts_sorted_end.index, hour_counts_sorted_end.values, color='red')
axes[1].set_xlabel('Hour')
axes[1].set_ylabel('Count')
axes[1].set_title('Count of Ended Trips for Each Hour')
axes[1].tick_params(axis='x') 
plt.sca(axes[1])
plt.xticks(hour_counts_sorted_end.index)

# Display the plots
plt.tight_layout()
plt.show()
```

- The same pattern was observed for arrivals.
- There was an increase in the number of trips recorded during arrivals at 18 and 19 o'clock compared to the trips registered during departures.

```python
# Define function for plotting horizontal bar chart with labels
def plot_horizontal_bar_with_labels(ax, data, color, title):
    ax.barh(np.arange(len(data)), data.values, color=color)
    ax.set_yticks(np.arange(len(data)))
    ax.set_yticklabels(data.index)
    ax.set_xlabel('Frequency')
    ax.set_title(title)
    ax.invert_yaxis()  # Invert y-axis to display highest frequency at the top
    ax.grid(axis='x')  # Add gridlines
    
    # Add labels in front of the bars
    for i, v in enumerate(data):
        ax.text(v + 0.2, i, str(v), color='black', va='center')

# Set up the figure and axes
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

# Plot top 10 male age distribution
plot_horizontal_bar_with_labels(axes[0], top10_age_male_dist, 'skyblue', 'Top 10 Male Age Distribution')

# Plot top 10 female age distribution
plot_horizontal_bar_with_labels(axes[1], top10_age_female_dist, 'pink', 'Top 10 Female Age Distribution')

plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```
