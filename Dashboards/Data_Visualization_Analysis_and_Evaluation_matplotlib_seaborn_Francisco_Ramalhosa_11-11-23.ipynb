````python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
````

````python


df = pd.DataFrame([
    ['Acidental', .09],
    ['Descuido', .24],
    ['Intencional', .01],
    ['Falso Alarme', .08],
    ['Infundada', .05],
    ['Indeterminada', .34],
    ['Curto Circuito', .19],
], columns=['causa', 'perc'])
````

````python
df['perc']=df['perc']*100
df = df.sort_values(by='perc', ascending=False)
df
````

````python
sns.barplot(x=df['perc'], y=df['causa'], color='blue')
ax = plt.gca()
for i, v in enumerate(df['perc']):
     ax.text(v, i, f'{v:.1f}%', color='k')
ax.set_title(f"Distribuição das causas de incêndios em edifícios.", loc='left')
ax.set_ylabel(' ')
ax.set_xlabel(' ')
ax.xaxis.set_label_coords(0.8, -0.1)
sns.despine(left=True, bottom=True)
ax.tick_params(axis='both', which='both', bottom=False, left=False, labelbottom=False)
ax.set_xlim(0, 40)
plt.show()
````

**R:**


- na minha opinião a escolha do tipo gráfico inadequada: um 'pie-chart' para esta situação em concreto não é a melhor opção. Existem 7 categorias ('causas') a serem representadas neste gráfico, a utilização do 'pie-chart' pode dificultar e gerar confusão na sua interpretação, uma vez que, devido ao considerável número de categorias, determinadas 'fatias' serão dificeis de diferenciar entre si em termos dos seus valores de percentagens e leitura dos 'labels';
- 'labels' indevidamente colocados: por se tratar de um 'pie-chart' com considerável número de categorias, as 'fatias' que representam 'causas' com menor distribuição percentual não tem o espaço suficiente para incluir no seu interior a descrição e valor da percentagem da 'causa' correspondente.  
Como consequencia, alguns dos 'labels' ficam representados no interior do gráfico circular enquanto que outros ficam representados no seu exterior. Uma possivel alternativa seria colocar apenas os valores das percentagens dentro de cada 'fatia' e fazer colocar a descrição de cada 'causa' do lado de fora de cada 'fatia', ou então, uma outra hipotese seria criar uma legenda que representasse cada uma das 'causas' com diferentes cores;  
- muitas cores diferentes – no caso do 'pie-chart' acaba por ser adequado a utilização de cores distintas, mas se por exemplo representassemos os dados através de um gráfico de barras ou colunas, a melhor opção seria utilizar a mesma cor para facilitar a leitura da visualização;
- 3D é inapropriado: uma eventual representação gráfica em 2D seria mais prática e simples de interpretar comparativamente à representação em 3D;
- título escolhido poderia ser mais adequado: o titulo atribuido poderia ser mais direto e dar uma melhor contextualização do tema abordado.  

Tendo em consideração os motivos apresentados, uma possível alternativa à representação gráfica dos dados poderia ser um gráfico de barras horizontal. Em que:
- as categorias estão identificadas por ordem decrescente de distribuição;
- os respetivos valores percentuais de cada categoria devidamente identificados, imediatamente à frente de cada uma das barras (através de rótulos);
- uma única cor utilizada para todas as 'causas' permite uma visualização mais 'limpa' e menos confusa do gráfico (nesta situação em concreto, não existe a necessidade de distinguir as categorias através de cores);
- os nomes das 'causas' identificados por meio dos rótulos do eixo vertical;
- título informativo e prático, permite a rápida contextualização da representação gráfica;
- marcadores e linhas dos eixos verticais e horizontais, e rótulos do eixo horizontal, opcionalmente omitidos com o objetivo de tornar a interpretação do gráfico o mais simples possível;

A meu ver, neste contexto, trata-se de uma boa prática, pois desta forma evita-se a desnecessária utilização adicional de uma legenda, dedicando assim, a totalidade do espaço existente à representação gráfica dos dados.

````python
cust_data = pd.read_csv(os.path.join('telco', 'WA_Fn-UseC_-Telco-Customer-Churn.csv') , skipinitialspace=True)
cust_data['SeniorCitizen'] = cust_data['SeniorCitizen'].astype(str)
cust_data=cust_data.dropna()
cust_data=cust_data[cust_data['tenure']<=12]
````

````python
# número de clientes séniores e clientes não séniores que fizeram contrato com a empresa.

total_clientes=cust_data['customerID'].count()
print('Total number of customers with 12 or fewer months of membership: ', total_clientes)
clientes_seniores_nao_seniores=cust_data.groupby('SeniorCitizen')['SeniorCitizen'].count()
print(clientes_seniores_nao_seniores)
perc_clientes_seniores_nao_seniores=clientes_seniores_nao_seniores/total_clientes*100
perc_clientes_seniores_nao_seniores.round(1)

labels = ['Not Senior', 'Senior']
data = perc_clientes_seniores_nao_seniores.round(1)
explode = (0, 0.05)
plt.figure(figsize=(8, 10))
plt.pie(data, explode=explode, labels=labels, autopct='%1.1f%%', startangle=145)
plt.title("Distribution of Non-Senior and Senior customers.")
plt.show()
````

````python
# número de clientes séniores e clientes não séniores por género.

clientes_seniores_nao_seniores_por_genero=cust_data.groupby(['SeniorCitizen', 'gender'])['gender'].count()
clientes_seniores_nao_seniores_por_genero=clientes_seniores_nao_seniores_por_genero/clientes_seniores_nao_seniores*100
clientes_seniores_nao_seniores_por_genero.round(1)
````

````python
# número de clientes séniores e clientes não séniores por tipo de contrato.

clientes_seniores_nao_seniores_por_contrato=cust_data.groupby(['SeniorCitizen', 'Contract'])['Contract'].count()
clientes_seniores_nao_seniores_por_contrato=clientes_seniores_nao_seniores_por_contrato/clientes_seniores_nao_seniores*100
clientes_seniores_nao_seniores_por_contrato.round(1)

df_plot = clientes_seniores_nao_seniores_por_contrato.unstack(level=0)
ax = df_plot.plot(kind='bar', figsize=(10, 6), colormap='viridis', width=0.8)
plt.title('Distribution of Non-Senior and Senior customers by type contract')
plt.xlabel('Contract type')
plt.ylabel('%')
plt.xticks(rotation=0, ha='center')  # Rótulos do eixo horizontal totalmente horizontais
ax.yaxis.set_label_coords(-0.06, 0.95)
ax.legend().set_visible(False)
plt.show()
````

````python
# número de clientes séniores e clientes não séniores que recebem fatura eletrónica.

clientes_seniores_nao_seniores_por_paperlessbilling=cust_data.groupby(['SeniorCitizen', 'PaperlessBilling'])['PaperlessBilling'].count()
clientes_seniores_nao_seniores_por_paperlessbilling=clientes_seniores_nao_seniores_por_paperlessbilling/clientes_seniores_nao_seniores*100
clientes_seniores_nao_seniores_por_paperlessbilling.round(1)

df_plot = clientes_seniores_nao_seniores_por_paperlessbilling.unstack(level=0)
ax = df_plot.plot(kind='bar', figsize=(10, 6), colormap='viridis', width=0.8)
plt.title('Distribution of Non-Senior and Senior customers who receive paperless billing')
plt.xlabel('Paperless billing?')
plt.ylabel('%')
plt.xticks(rotation=0, ha='center')  # Rótulos do eixo horizontal totalmente horizontais
ax.yaxis.set_label_coords(-0.06, 0.95)
plt.legend(['Not Senior', 'Senior'], title='Client')
ax.legend().set_visible(False)
plt.show()
````

````python
# número de clientes séniores e clientes não séniores por tipo de meio de pagamento.

clientes_seniores_nao_seniores_por_paymentmethod=cust_data.groupby(['SeniorCitizen', 'PaymentMethod'])['PaymentMethod'].count()
clientes_seniores_nao_seniores_por_paymentmethod=clientes_seniores_nao_seniores_por_paymentmethod/clientes_seniores_nao_seniores*100
clientes_seniores_nao_seniores_por_paymentmethod.round(1)

df_plot = clientes_seniores_nao_seniores_por_paymentmethod.unstack(level=0)
ax = df_plot.plot(kind='barh', figsize=(10, 6), colormap='viridis', width=0.8)
plt.title('Distribution of Senior and Non-Senior Customers through payment method')
plt.xlabel('%')
plt.ylabel('')
plt.xticks(rotation=0, ha='center')  # Rótulos do eixo horizontal totalmente horizontais
plt.legend(['Not Senior', 'Senior'], title='Client')
ax.tick_params(axis='y', which='both', bottom=True, left=False, labelbottom=True)
ax.legend().set_visible(False)
plt.show()
````
