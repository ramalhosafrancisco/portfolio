## Importação de bibliotecas e criação do Dataframe

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
df = pd.read_csv('IMDB-Movie-Data.csv', index_col='Rank')
df.rename(columns={
    'Runtime (Minutes)': 'Runtime',
    'Revenue (Millions)': 'Revenue'
}, inplace=True)
df.columns = [col.lower() for col in df.columns]
df
```

## Contexto

Além da classificação da IMDB ('rating') existem também outras variáveis a serem consideradas, nomeadamente: 'title', 'genre', 'description', 'director', 'actors', 'year', 'runtime', 'votes', 'revenue' e 'metascore'.

## Gráficos de visualização univariada

### Gráfico de barras vertical que mede a distribuição do numero de lançamento de filmes entre 2006 e 2016

```python
plt.figure(figsize=(10, 6))
sns.countplot(x=df["year"], color='grey')
plt.title('Distribuição de Filmes ao Longo dos Anos')
plt.xlabel('ano')
plt.ylabel('Número de Filmes')
plt.show()
```

O gráfico tem como objetivo determinar quais os anos em que foram lançados o maior e o menor número de filmes que constam no Dataset fornecido.
Através da visualização do gráfico é possivel verificar que existe uma tendencia crescente no aumento do lançamneto de novos filmes no decorrer dos anos. A partir de 2013 essa diferença é mais acentuada, atingindo o pico do ano de 2016, onde foram lançados quase o dobro dos filmes em relaçao ao ano anterior. Em 2006 verificou-se o menor número de filmes produzidos e entre 2007 e 2012 a média de filmes realizados manteve-se praticamente constante.

### Gráfico de barras horizontal com a distribuição de filmes por classificação qualitativa 
#### (Excelente, Muito Bom, Bom, Mediano, Mau)

```python
def get_new_rating(old_rating):
    if old_rating >= 9:
        return 'Excelente'
    elif 8<old_rating<9:
        return 'Muito Bom'
    elif 7<old_rating<8:
        return 'Bom'
    elif 5<old_rating<7:
        return 'Mediano'
    else:
        return 'Mau'

df['new rating'] = df['rating'].apply(get_new_rating)
df.info()
```

```python
plt.figure(figsize=(10, 6))
sns.catplot(data=df,
            y='new rating',
            kind='count',
            order=df['new rating'].value_counts().index,)
plt.title('Distribuição de Filmes por classificação nominal')
plt.xlabel('Número de Filmes')
plt.ylabel('')
plt.show()
```

```python
df[df['new rating'] == 'Excelente']
```

Com o objetivo de facilitar a interpretação do DF, foi feita uma nova atribuição da classificação qualitativa nominal com base na pontuação média atribuida e calculada pela IMDB para cada filme. Neste caso, as novas classificação são: excelente (que corresponde à classificação IMDB entre 9 e 10); muito bom (que corresponde à classificação IMDB entre 8 e 9); bom (que corresponde à classificação IMDB entre 7 e 8); mediano (que corresponde à classificação IMDB entre 5 e 7) e mau (que corresponde à classificação IMDB abaixo de 5).  
Por interpretação do gráfico verifica-se que a maioria dos filmes que foram lançados tem uma nova classicação 'mediana', seguindo-se de 'bom' e 'mau'. Os filmes com classificação 'muito bom' e 'excelente' são os menos representados neste gráfico (de salientar ainda que existe apenas um único filme com classificação 'excelente': 'The Dark Knight').

## Gráfico de visualização bivariada

### Grafico de dispersão que faz a relação entre ano de lançamento, receita e classificação nominal dos filmes

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='year', y='revenue', data=df, hue='new rating', palette='deep')
plt.title('Filmes: Ano de Lançamento vs Receita vs Classificação Nominal')
plt.xlabel('Ano de lançamento')
plt.ylabel('Receita (em milhões de dólares)')
plt.legend(title='Classificação Nominal')
plt.show()
```

As relações que podem ser observadas no gráfico incluem como a receita dos filmes varia ao longo dos anos e como essa variação difere para diferentes classificações nominais.  
É possivel verificar que a maioria dos filmes com mais do que 400M dólares de receitas tem na sua esmagadora maioria classificação nominal maior que mediano (há exceção de dois filmes). O filme mais lucrativo é também o filme com melhor classificação. Existe um filme lançado em 2007 com classificação 'Muito Bom' mas curiosamente com uma das receitas mais baixas observadas em todo o gráfico. A maioria dos filmes tem receitas inferiores a 200 M dólares. Não parece existeir diferenças significativas no que se refere à distribuição das receitas nos diferentes anos analisados.

## Principais 'Insights'

1. Tendência de lançamento de filmes:  

A tendência geral mostra um aumento constante no lançamento de filmes ao longo dos anos, com um pico significativo em 2016, indicando um crescimento na produção cinematográfica.  

2. Ano de 2006 como ano com menor número de lançamentos:  

O ano de 2006 destaca-se como o ano com o menor número de filmes lançados, enquanto entre 2007 e 2012 a média de filmes realizados manteve-se praticamente constante.  

3. Distribuição de classificações nominais:  

A maioria dos filmes lançados tem uma classificação qualitativa 'mediana', seguida por 'bom' e 'mau'. As classificações 'muito bom' e 'excelente' são menos comuns, com destaque para apenas um filme com classificação 'excelente' ('The Dark Knight').  

4. Relação entre receita e classificação nominal:  

Filmes com receitas superiores a 400 milhões de dólares tendem a ter classificações nominais mais altas (com exceção de raros casos), indicando uma possível correlação positiva entre receita e classificação nominal.  

5. Variação nas receitas ao longo dos anos:  

Não há diferenças significativas na distribuição dos valores das receitas ao longo dos anos, indicando que, apesar do aumento na produção de filmes, a variação dos valores nas receitas não é muito evidente.  
A maioria dos filmes, independentemente do ano de lançamento, tem receitas inferiores a 200 milhões de dólares, sugerindo uma certa estabilidade nas faixas de receitas ao longo do período analisado.
