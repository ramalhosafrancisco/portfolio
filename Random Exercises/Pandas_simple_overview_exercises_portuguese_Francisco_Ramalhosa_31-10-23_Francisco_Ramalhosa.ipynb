### Qual a diferença entre um atributo e uma função/método no 'pandas'?

Atributos são características associadas a um objeto Pandas (como um DataFrame ou uma Série). Fornecem informações sobre o objeto, como tamanho, forma, tipo de dados, etc..Pode-se utilizar os atributos diretamente, sem usar parênteses. Exemplos de atributos: .shape, .dtype, .columns, .index, entre outros.
Funções (ou métodos) são ações que se podem realizar nos objetos Pandas, como por exemplo, cálculos, filtragem, ordenação, agregação, entre outros. São chamadas utilizando parênteses, e geralmente modificam o objeto original. Exemplos de funções/métodos: head(), tail(), groupby(), etc..

### O que é uma série e um dataframe em 'pandas'? Como se relacionam?

Séries e DataFrames(DF) são estruturas de dados que são usadas para armazenar, organizar e manipular dados.
Os DataFrames permitem organizar esses dados de maneira tabular, onde diferentes colunas podem conter diferentes variáveis e tipos de dados. As Séries são usadas para representar uma única coluna ou uma única dimensão de dados, por isso os DF são bidimenssionais enquanto que as Séries são unidimenssionais.
Cada coluna de um DF é uma Séries. As Séries podem conter dados de qualquer tipo, números, strings, datas, etc.
DataFrames e Séries têm um índice associado para rotular cada uma das linhas.
